```python
#Step 9:- Which was the most ordered item?
most_ordered = chipo.groupby("item_name")["quantity"].sum().sort_values(ascending=False).head(1)
print("Most ordered item:\n",most_ordered)
```

```python
#10:- For the most-ordered item, how many items were ordered?
most_ordered_item = most_ordered.index[0]
most_ordered_quantity = most_ordered.values[0]
print(f"The most ordered item is '{most_ordered_item}' with {most_ordered_quantity} orders.")
```

```python
#11:-What was the mostv ordered item in the choice_desription column?
most_ordered_choice = (
    chipo.dropna(subset=["choice_description"])
         .groupby("choice_description")["quantity"]
         .sum()
         .sort_values(ascending=False)
         .head(1)
)

q11 = plt.bar(most_ordered_choice.index, most_ordered_choice.values, color="orange", edgecolor="black", width=0.8)
# bars = plt.bar(grade_counts.index, grade_counts.values, color="orange", edgecolor="black")

plt.title("Most ordered choice_description item")
plt.xlabel("item")
plt.ylabel(" count")
plt.show()


print("Most ordered choice_description item:")
print(most_ordered_choice)
```

```python
#Step 12:- How many items were ordered?
total_items_ordered = chipo["quantity"].sum()
print("Total number of items ordered:", total_items_ordered)
```

```python
#Step 13:- Turn the item price into float.

chipo["item_price"] = chipo["item_price"].str.replace("$", "", regex=False).astype(float)

chipo["item_price"].head()
```

```python
 #Step 13.b.Create a lambda function and change the type of item price

chipo["item_price"] = chipo["item_price"].apply(lambda x: float(str(x).replace("$", "")))

chipo["item_price"].head()
```

```python
#Step 13.c.Check the item price type
print("Data type of 'item_price':", chipo["item_price"].dtype)
```

```python
#Step 14:- How much was the revenue for the period in the dataset?

chipo["revenue"] = chipo["quantity"] * chipo["item_price"]
total_revenue = chipo["revenue"].sum()
print(f"Total revenue for the period: ${total_revenue:,.2f}")
```

```python
#Step 15:- How many oredrs were made in the period?

total_orders = chipo["order_id"].nunique()
print("Total number of orders in the period:", total_orders)
```

```python
#Step 16:-What is the average revenue amount per order?

avg_revenue_per_order = chipo.groupby("order_id")["revenue"].sum().mean()
print(f"Average revenue per order: ${avg_revenue_per_order:,.2f}")
```

```python
#Step 17:- How many different items are sold?

num_unique_items = chipo["item_name"].nunique()
print("Number of different items sold:", num_unique_items)
```

#Data Visualization

```python
# 1.Create a bar chart of top 10 most-ordered items.

top_items = (
    data.groupby("item_name")["quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_items.values, y=top_items.index, hue=top_items.index, palette="viridis", dodge=False, legend=False
)

plt.title("Top 10 Most Ordered Items")
plt.xlabel("Total Quantity Ordered")
plt.ylabel("Item Name")

plt.show()
```

```python
#2.create a pie chart showing the proportion of different choice_description categories.

# Step 1: Count categories
# Step 2: Keep top 3, group the rest into "Other"
# Step 3: Plot pie chart

import matplotlib.pyplot as plt
import pandas as pd

choice_counts = data['choice_description'].dropna().value_counts()

top_n = 5
choice_counts_top = choice_counts[:top_n]
choice_counts_top["Other"] = choice_counts[top_n:].sum()

plt.figure(figsize=(8, 8))
plt.pie(choice_counts_top.values, labels=choice_counts_top.index, autopct='%1.1f%%',
        startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title("Top 5 Choice Descriptions (Others grouped)")
plt.show()
```

```python
# TOP 10 IN BARH PLOT (for reference )
plt.figure(figsize=(10, 8))
choice_counts[:10].plot(kind='barh')
plt.title("Top 15 Choice Descriptions")
plt.xlabel("Count")
plt.ylabel("Choice Description")
plt.gca().invert_yaxis()  # so largest is on top
plt.show()
```

```python
# 3. Identify high revenue generating items

# Step 1: Clean item_price (remove "$" and convert to float) - This step is already done in previous cells for the
# 'chipo' dataframe, and the 'data' dataframe's 'item_price' is already float as seen in the last successful execution of this cell block.
# Step 2: Treat item_price as per-row revenue (already includes quantity)
# Step 3: Aggregate revenue by item
# Step 4: Plot top 10 revenue items
data["revenue"] = data["item_price"] * data["quantity"]

revenue_items = (
    data.groupby("item_name")["revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))
sns.barplot(x=revenue_items.values, y=revenue_items.index, palette="magma")

plt.title("Top 10 Revenue-Generating Items")
plt.xlabel("Total Revenue ($)")
plt.ylabel("Item Name")
plt.show()

#display(revenue_items)
```

```python
# Suggest business insights from your analysis (2-3 key points)
# Business Insights from Chipotle Dataset

display(revenue_items)

#1. Top revenue items
top_revenue = (
    data.groupby("item_name")["revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

# 2.Top ordered items (by quantity)
top_orders = (
    data.groupby("item_name")["quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

# 3.Choice variety
unique_choices = data["choice_description"].nunique()

print(" Business Insights")
print(f"1 Core items drive most revenue: \n{top_revenue}\n")
print(f"2 Most frequently ordered items: \n{top_orders}\n")
print(f"3 Customers value customization â†’ {unique_choices} unique choice descriptions.\n")
```

```python
#better to visualise to compare

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Histogram for revenue
axes[0].hist(data["revenue"], bins=30, color="skyblue", edgecolor="black")
axes[0].set_title("Revenue Distribution")
axes[0].set_xlabel("Revenue")
axes[0].set_ylabel("Frequency")

# 2. Histogram for quantity
axes[1].hist(data["quantity"], bins=30, color="lightgreen", edgecolor="black")
axes[1].set_title("Quantity Distribution")
axes[1].set_xlabel("Quantity")
axes[1].set_ylabel("Frequency")

# 3. Histogram for choice variety
choice_counts = data.groupby("item_name")["choice_description"].nunique()
axes[2].hist(choice_counts, bins=15, color="salmon", edgecolor="black")
axes[2].set_title("Choice Variety per Item")
axes[2].set_xlabel("Unique Choices")
axes[2].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
